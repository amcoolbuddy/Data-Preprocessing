# How to deal with missing values using scikit learn

### First deal with the variables which needs to be replaced by mean.

```python
from sklearn.impute import SimpleImputer
imputer_mean= SimpleImputer(missing_values=np.nan, strategy= 'mean')
df[['normalized-losses','stroke','bore','horsepower','peak-rpm']]=imputer_mean.fit_transform(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']]).astype('object')
```

```python
df.head(10)
```

### Second deal with the variable which needs to be replaced by most frequent value

```python
imputer_most_freq = SimpleImputer(missing_values= np.nan, strategy='most_frequent')
df[['num-of-doors']]=imputer_most_freq.fit_transform(df[['num-of-doors']]).astype('object')
```

```python
df.head()
```

```python
df.info()
```

### Third drop the row having price value missing

```python
# Simple drop the whole row with nan in price column
df.dropna(subset=['price'],axis=0,inplace=True)

# Reset index because we dropped some of the rows , hence we have to fix the sequence of indices now.
df.reset_index(drop= True, inplace =True)
```

## 2.Data Binning

- Binning is the process of transforming continuous numerical variables into discrete categorical 'bins',for grouped analysis. 
- Example :-
- In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it has 57 unique values. What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? Can we rearrange them into three bins to simplify analysis?
We will use the pandas method 'cut' to segment the 'horsepower' column into 3 bins

```python
## Lets first convert horsepower datatype to int or float.
df['horsepower'] =df['horsepower'].astype(float)
df['horsepower'].dtypes
```

```python
# Let's visualize the horsepower data in histogram
import matplotlib.pyplot as plt
plt.hist(df['horsepower'])
plt.xlabel("Horsepower Values")
plt.ylabel("Frequency count")
```

```python
# Using linspace-> Linspace is a numpy method with the help of which we can equally distribute our data.
np.linspace(1,10,4)
```

```python
## Create bins
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
## Create group names
group_names =['Low','Medium','High']
```

```python
# Perform binning
df['horsepower-binned'] = pd.cut(df['horsepower'], bins,labels=group_names, include_lowest=True)
# bins -> for giving range
# labels -> giving name to those bins
# include lowest-> to include 48 also i.e. the lowest value
df[['horsepower','horsepower-binned']].head(20)
```

```python
# See the frequency of each bin with value_counts()
df['horsepower-binned'].value_counts()
```

```python
# Lets visualize again with bins.
plt.bar(group_names, df['horsepower-binned'].value_counts())
plt.xlabel("horsepower")
plt.ylabel('count')
```
